Program 1: Mean,Mode,Median using the NumPy,Matplotlib and SciPy libraries.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# User input
n = int(input("Enter number of elements: "))
print("Enter the elements:")

data = []
for i in range(n):
    data.append(float(input()))

# Convert to NumPy array
arr = np.array(data)

# Mean, Median, Mode
mean_value = np.mean(arr)
median_value = np.median(arr)
mode_value = stats.mode(arr, keepdims=True)

# Display results
print("\nResults:")
print("Mean   =", mean_value)
print("Median =", median_value)
print("Mode   =", mode_value.mode[0])

# Plotting
plt.hist(arr, bins=10)
plt.axvline(mean_value)
plt.axvline(median_value)
plt.axvline(mode_value.mode[0])
plt.title("Histogram with Mean, Median, and Mode")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

Program 2:Variance,Standard Deviation using the NumPy,Matplotlib and SciPy libraries.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# User input
n = int(input("Enter number of elements: "))
print("Enter the elements:")

data = []
for i in range(n):
    data.append(float(input()))

# Convert to NumPy array
arr = np.array(data)

# Variance and Standard Deviation
variance_np = np.var(arr)
std_dev_np = np.std(arr)

variance_sp = stats.tvar(arr)
std_dev_sp = stats.tstd(arr)

# Display results
print("\nUsing NumPy:")
print("Variance =", variance_np)
print("Standard Deviation =", std_dev_np)

print("\nUsing SciPy:")
print("Variance =", variance_sp)
print("Standard Deviation =", std_dev_sp)

# Plotting
plt.hist(arr, bins=10)
plt.axvline(np.mean(arr))
plt.title("Histogram with Mean")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```
